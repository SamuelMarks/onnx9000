# Tutorial: Building Custom ONNX Ops with onnx9000.script

In this tutorial, we will use `@onnx9000.script` to build custom operations cleanly, without relying on `protobuf`.

```
from onnx9000.script import op, script

op.set_target_opset(18)


@script
def swish_activation(x):
    return x * op.Sigmoid(x)


builder = swish_activation.to_builder()
builder.set_metadata(domain="custom.domain")
model = builder.to_onnx()
print("Custom op nodes:", [n.op_type for n in model.graph.node])
```
